# Functions

1. What is the difference between a function and a method in Python?
    - A function is a piece of reusable code you can call anywhere, defined with def.A method is like a function but belongs to a class or object.Methods can use self to work with an object's data,while functions work independently without being tied to objects.

2. Explain the concept of function arguments and parameters in Python?
    - In Python, parameters are placeholders in a function definition (e.g.,`def greet(name):`).Arguments are the actual values you pass when calling the function (e.g., `greet("Alice")`).Parameters define what the function expects, and arguments provide those values when the function runs.Together,they make functions flexible and reusable.

3. What are the different ways to define and call a function in Python?
    - In Python, you define a function using "def" with a name and parameters (optional). Call it by writing its name with parentheses, passing arguments if needed.Functions can be regular,lambda (short inline functions), or nested (inside another function),offering flexibility for different programming needs.

4. What is the purpose of the `return` statement in a Python function?
    - The `return` statement in a Python function sends a value back to where the function was called. It allows the function to produce a result that can be used later. Without `return`,the function only runs its code but doesn’t give anything back to the caller.

5. What are iterators in Python and how do they differ from iterables?
    - An iterable is an object you can loop through (like a list or string).An iterator is an object that produces items one at a time using `next()`.Iterables use `iter()` to create an iterator.Iterators remember their position, while iterables can start fresh every time.

6. Explain the concept of generators in Python and how they are defined?
    - Generators are special functions in Python that produce values one at a time using "yield" instead of "return". They pause and resume where they left off, making them memory-efficient.Define them like functions but use "yield". Useful for large data or infinite sequences without storing everything in memory.

7. What are the advantages of using generators over regular functions?
    - Generators are better than regular functions for handling large data or infinite sequences because they don’t store all results in memory. They create values one at a time, saving memory and improving performance.Generators are also easier to write and manage compared to manually creating iterators.

8. What is a lambda function in Python and when is it typically used?
    - A lambda function is a short,one-line function without a name, defined using the "lambda" keyword. It’s used for simple tasks,like quick calculations or sorting, where defining a full function is unnecessary. Lambdas are commonly used with functions like "map", "filter", and "sort" for concise,temporary logic.

9. Explain the purpose and usage of the `map()` function in Python?
    - The `map()` function applies a given function to each item in an iterable (like a list) and returns a map object.It’s used to transform or process data quickly without needing a loop.You can convert the result to a list or another iterable for further use.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
    - `map()` applies a function to each item in an iterable and returns a new iterable.
`filter()` filters elements based on a condition, returning only the ones that meet it.
`reduce()` (from `functools`) repeatedly applies a function to reduce the iterable to a single value.

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.
  

```python
def sum_even(numbers):
    return sum(num for num in numbers if num % 2==0)

numbers=[1,2,3,4,5,6]
result = sum_even(numbers)
print(result)
```

4. . Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def prime(number):
  if number<2:
    return False
  for i in range (2, int(number**0.5)+1):
    if number%i == 0:
      return False
    return True
  for num in range(1, 201):
    if prime(num):
        print(num, "is prime")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        self.count += 1
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        return result


```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple

```python
tuples_list = [(1, 3), (4, 1), (2, 2), (5, 0)]

sorted_list = sorted(tuples_list, key=lambda x: x[1])

print(sorted_list)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

```python
celsius_temps = [0, 20, 30, 40]
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

print(fahrenheit_temps)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
input_string = "Hello World"

result = ''.join(filter(lambda char: char.lower() not in 'aeiou', input_string))

print(result)
```

11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

```python
orders = [
    ["34587", "Learning Python, Mark Lutz", 4, 40.95],
    ["98762", "Programming Python, Mark Lutz", 5, 56.80],
    ["77226", "Head First Python, Paul Barry", 3, 32.95],
    ["88112", "Einführung in Python3, Bernd Klein", 3, 24.99],
]

def calculate_order_total(order):
    order_number = order[0]
    quantity = order[2]
    price_per_item = order[3]
    total = quantity * price_per_item
    if total < 100:
        total += 10
    return (order_number, total)

# Solution using map and lambda
result = list(map(lambda order: (order[0], (order[2] * order[3]) + 10 if (order[2] * order[3]) < 100 else (order[2] * order[3])), orders))

print(result)
```
